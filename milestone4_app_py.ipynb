```python
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime

# Page config
st.set_page_config(
    page_title="ClimateScope - Global Weather Dashboard",
    layout="wide",
    page_icon="🌍"
)

st.title("🌍 ClimateScope")
st.subheader("Global Weather Trends & Extreme Events Analysis")
st.markdown("Interactive dashboard for exploring temperature, humidity, wind, and precipitation patterns.")

# Load Data
@st.cache_data
def load_data():
    # Change path if needed
    df = pd.read_csv("/content/global-weather-repository/GlobalWeatherRepository.csv")
    df['date'] = pd.to_datetime(df['last_updated'], errors='coerce')
    return df

# Call load_data() outside the function definition to make 'df' available globally
df = load_data()

# Sidebar Filters
st.sidebar.header("Filters")

# Country filter
countries = sorted(df['country'].unique()) if 'country' in df.columns else []
selected_countries = st.sidebar.multiselect(
    "Select Countries",
    options=countries,
    default=countries[:5] if countries else []
)

# Metric selection
metric_options = {
    "Temperature (°C)": "temperature_celsius",
    "Humidity (%)": "humidity",
    "Wind Speed (km/h)": "wind_kph",
    "Precipitation (mm)": "precip_mm" # Changed from 'precipitation_mm'
}

selected_metric_label = st.sidebar.selectbox("Select Metric", options=list(metric_options.keys()))
selected_metric = metric_options[selected_metric_label]

# Initialize date_range to None, in case 'date' column is missing
date_range = None

# Date range
if 'date' in df.columns:
    min_date = df['date'].min()
    max_date = df['date'].max()
    date_range_input = st.sidebar.date_input(
        "Date Range",
        value=(min_date, max_date),
        min_value=min_date,
        max_value=max_date
    )
    # Ensure date_range is a list of datetime objects for consistent handling
    if date_range_input and len(date_range_input) == 2:
        date_range = [pd.to_datetime(d) for d in date_range_input]
else:
    # Provide a warning if date filtering is not possible
    st.sidebar.warning("No 'date' column found in data for date filtering.")

# Apply filters (This block was causing the IndentationError)
filtered_df = df.copy()

if selected_countries:
    filtered_df = filtered_df[filtered_df['country'].isin(selected_countries)]

# Apply date filter only if 'date' column exists and date_range is valid
if 'date' in filtered_df.columns and date_range is not None and len(date_range) == 2:
    filtered_df = filtered_df[
        (filtered_df['date'] >= date_range[0]) &
        (filtered_df['date'] <= date_range[1])
    ]

# KPI Cards
st.subheader("Key Climate Insights")
col1, col2, col3, col4 = st.columns(4)

col1.metric("Avg Temperature", f"{filtered_df['temperature_celsius'].mean():.1f} °C" if 'temperature_celsius' in filtered_df.columns else "N/A")
col2.metric("Avg Humidity", f"{filtered_df['humidity'].mean():.1f} %" if 'humidity' in filtered_df.columns else "N/A")
col3.metric("Avg Wind Speed", f"{filtered_df['wind_kph'].mean():.1f} km/h" if 'wind_kph' in filtered_df.columns else "N/A")
col4.metric("Avg Precipitation", f"{filtered_df['precip_mm'].mean():.1f} mm" if 'precip_mm' in filtered_df.columns else "N/A")
```

```python
# Tabs (matching common structure in such ClimateScope dashboards)
tab1, tab2, tab3, tab4, tab5 = st.tabs([
    "📈 Overview Trends",
    "🌡️ Temperature Analysis",
    "🌧️ Precipitation & Wind",
    "⚠️ Extreme Events",
    "📊 Correlations"
])

with tab1:
    st.subheader(f"{selected_metric_label} Trend Over Time")
    fig = px.line(filtered_df, x="date", y=selected_metric, color="country",
                  title=f"{selected_metric_label} Trend by Country")
    st.plotly_chart(fig, use_container_width=True)

with tab2:
    st.subheader("Temperature Distribution")
    fig_temp = px.histogram(filtered_df, x="temperature_celsius", nbins=50, color="country")
    st.plotly_chart(fig_temp, use_container_width=True)

    st.subheader("Temperature Distribution")
    fig_temp = px.histogram(filtered_df, x="temperature_celsius", nbins=50, color="country")
    st.plotly_chart(fig_temp, use_container_width=True)

    st.subheader("Country-wise Average Temperature")
    avg_temp = filtered_df.groupby("country")["temperature_celsius"].mean().reset_index()
    fig_bar = px.bar(avg_temp, x="country", y="temperature_celsius", color="temperature_celsius")
    st.plotly_chart(fig_bar, use_container_width=True)

with tab3:
    st.subheader("Precipitation vs Wind Speed")
    fig_scatter = px.scatter(filtered_df, x="wind_kph", y="precip_mm", # Changed from 'precipitation_mm'
                             color="country", size=filtered_df["temperature_celsius"].abs()) # Use absolute value for size
    st.plotly_chart(fig_scatter, use_container_width=True)

with tab4:
    st.subheader("Extreme Events Detection")

    # Extreme Heat
    heat_threshold = filtered_df["temperature_celsius"].quantile(0.95)
    extreme_heat = filtered_df[filtered_df["temperature_celsius"] > heat_threshold]
    st.metric("Extreme Heat Events", len(extreme_heat))

with tab5:
    st.subheader("Correlation Between Climate Variables")
    numeric_df = filtered_df.select_dtypes(include=['float64', 'int64'])
    corr = numeric_df.corr()
    fig_corr = px.imshow(corr, text_auto=True, color_continuous_scale='RdBu_r')
    st.plotly_chart(fig_corr, use_container_width=True)
```
